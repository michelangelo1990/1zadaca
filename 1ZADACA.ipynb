###NEEDLEMAN-WUNSCH

Needleman-Wunsch algoritam je povijesno prva primjena dinamičkog programiranja na usporedbu bioloških sekvenci. Neka su $A$ i $B$ dvije biološke sekvence duljina $n$ i $m$.$S$ je supstitucijska matrica, $S(a,b)$ označava cijenu zamjene 
znaka $a$ sa znakom $b$. Koristit demo linearni model ocjene procjepa,$S(a,-)=S(-,a)=d$ , gdje je $d$ cijena 
mutacije ispuštanja odnosno umetanja. Poravnanjem sekvenci $A$ i $B$ nazivamo dvije nove biološke sekvence 
jednake duljine, $A'$ i $B'$ nastale tako da su u sekvence i uneseni razmaci (procjepi) na određenim mjestima. 
Vrijednost poravnanja $A'$ i $B'$ definiramo kao: 

$$F(A',B')=\sum_{i=1}^{|A'|=|B'|} S(A'[i],B'[i])$$

Optimalno poravnanje je najveća vrijednost od tri moguća načina konstrukcije:

$$P(i,j)=max \begin{cases}
P(i,j-1)+S(-,B[j]) \\
P(i-1,j)+S(A[i],-) , i,j \ge 1\\
P(i-1,j-1)+S(A[i],B[j])
\end{cases}$$

uz 
$$\begin{align} P(0,j)&=P(0,j-1)+S(-,B[j]), j\ge 1 \\
P(i,0)&=P(i-1,0)+S(A[i],-), i\ge 1\\
P(0,0)&=0.
\end{align}$$


```python
#učitavamo BLOSUM50 supstitucijsku matricu
from numpy import *

def ucitaj(ime):
    Sm=loadtxt(ime)
    S=dict()
    t = ['A', 'R', 'N', 'D', 'C', 'Q', 'E', 'G', 'H', 'I', 'L', 'K', 'M', 'F', 'P', 'S', 'T', 'W', 'Y', 'V'];
    for i in range(len(t)):
        redak = t[i]
        for j in range(len(t)):
            stupac=t[j]
            S[redak+stupac]=Sm[i,j]
    return S


#racunamo matricu prijelaza P
def needleman_wunsch(seq1, seq2, gap):
    r= len(seq1) + 1
    s= len(seq2) + 1
    P=zeros((r,s),float)
    supstitucijska_matrica=ucitaj("BLOSUM50.txt")
    for i in range(0, r):
        P[i,0] = i * gap
 
    for j in range(0, s):
        P[0,j] = j * gap
 
    for i in range(1, r):
        for j in range(1, s):
            match = P[i - 1,j - 1] + supstitucijska_matrica[seq1[i - 1] + seq2[j - 1]]
            delete = P[i - 1,j] + gap
            insert = P[i,j - 1] + gap
            P[i,j] = max(match, delete, insert)
 
    return trackback(P, seq1, seq2, gap, supstitucijska_matrica)

def trackback(P, seq1, seq2, gap, matrica):
    alignedSeq1 = ''
    alignedSeq2 = ''
    i = len(seq1)
    j = len(seq2)
    while i > 0 and j > 0:
        score = P[i,j]
        diagScore = P[i - 1,j - 1]
        upScore = P[i,j - 1]
        leftScore = P[i -1,j]
        if score == diagScore + matrica[seq1[i - 1] + seq2[j - 1]]:
            alignedSeq1 = seq1[i - 1] + alignedSeq1
            alignedSeq2 = seq2[j - 1] + alignedSeq2
            i -= 1
            j -= 1
        elif score == leftScore + gap:
            alignedSeq1 = seq1[i - 1] + alignedSeq1
            alignedSeq2 = '-' + alignedSeq2
            i -= 1
        elif score == upScore + gap:
            alignedSeq1 = '-' + alignedSeq1
            alignedSeq2 = seq2[j - 1] + alignedSeq2
            j -= 1
        else:
            stderr.write('Greska!')
    while i > 0:
        alignedSeq1 = seq1[i - 1] + alignedSeq1
        alignedSeq2 = '-' + alignedSeq2
        i -= 1
 
    while j > 0:
        alignedSeq1 = '-' + alignedSeq1
        alignedSeq2 = seq2[j - 1] + alignedSeq2
        j -= 1
 
    return alignedSeq1, alignedSeq2

seq1 = 'HEAGAWGHEE'
seq2 = 'PAWHEAE'
prvi, drugi=needleman_wunsch(seq1,seq2,-8)
prvi,drugi
```

U teoriji informacije, Hammingova udaljenost dvaju stringova jednake duljine je broj pozicija u kojima su odgovarajući simboli različiti. Drugim riječima, mjeri minimalni broj supstitucija potreban za promjenu jednog u drugo, ili broj grešaka koji je jedan string transformirao u drugi.

```python

from __future__ import division

r1 = "ACCCAGGTTAACGGTGACCAGGTACCAGAAGGGTACCAGGTAGGACACACGGGGATTAA"
q1 = "ACCGAGGTTAACGGTGACCAGGTACCAGAAGGGTACCAGGTAGGAGACACGGCGATTAA"
q2 = "TTCCAGGTAAACGGTGACCAGGTACCAGTTGCGTTTGTTGTAGGAGACACGGGGACCCA"

def hamming_distance(s1, s2):
    
    differences = 0
    for c1, c2 in zip(s1, s2):
        if c1 != c2:
            differences += 1
    return differences

hamming_distance(r1, q1), hamming_distance(r1, q2), hamming_distance(first, second)
```

#PROGRAM ZA UMETANJE I EKSTRAKCIJU VODENOG ŽIGA U SLIKAMA



Umetanje vodenog žiga je postupak koji zahtijeva da se slika prikaže kao matrica te se potraži SVD dekompozicija matrice, da bismo dobili samu jezgru slike a nakon toga se ona spaja s drugom slikom u svrhu dobivanja žigosane slike. 

###SVD

SVD dekompozicija odnosno _Singular value decomposition_, je dekompozicija matrice $X$ reda $m\times n$ takva da vrijedi $X=U\Sigma V^T$ gdje su $U$ i $V$ redom $m\times m$ i $n\times n$ unitarne matrice, a $\Sigma$ dijagonalna $m\times n$ s nenegativnim singularnim vrijednostima. Da takav rastav postoji govori nam slijedeći teorem.
Neka je $X\in M_mn(\mathbb{R})$. Tada postoje unitarne matrice $U\in M_m$ i $V\in M_n$ i dijagonalna matrica $ \Sigma $ s elementima $\sigma_i$, gdje su $\sigma_i\ge 0$, $i=1,\dots , min(m,n)$, takve da je $ X=U\Sigma V^T  $. Bez smanjenja općenitosti možemo pretpostaviti da za dijagonalnu matricu $\Sigma$ vrijedi $\sigma_1\ge \sigma_2\ge \dots \ge \sigma_{min(m,n)}$

Učitajmo slike koje će koristiti kao cover odnosno original koji želimo žigosati, te žig i prikažimo cover:

```python

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from PIL import Image


donp = Image.open('don.bmp')
don = donp.convert('LA')
plt.figure(figsize=(9, 6))
plt.imshow(don);
lenap = Image.open('lena.bmp')
lena = lenap.convert('LA')

```

Slijedeći blok koda prebacuje slike u matrice i ponovno prikazuje sliku, no ovaj puta žig:

```python
donmat = np.array(list(don.getdata(band=0)), float)
donmat.shape = (don.size[1], don.size[0])
donmat = np.matrix(donmat)

lenamat = np.array(list(lena.getdata(band=0)), float)
lenamat.shape = (lena.size[1], lena.size[0])
lenamat = np.matrix(lenamat)

plt.figure(figsize=(9,6))
plt.imshow(lenamat, cmap='gray');
```

```python
U_orig,S_orig,V_orig = svd(donmat)
shape(S_orig)
```

Postupak ide tako da izračunamo SVD dekompoziciju matričnog prikaza slike koju želimo žigosati te matricu singularnih vrijednosti malo promijenimo tako da iskoristimo cijelu matricu našeg žiga.

```python
alpha=1/255.
S_don=diagsvd(S_orig,512,512)
Uw,Sw,Vw=svd(np.matrix(S_don+alpha*lenamat))
SW=diagsvd(Sw,512,512)
```

Spojimo to sve natrag tako da uzmemo $U$ i $V$ originalne slike te $\Sigma$ iz SVD dekompozicije promijenjene matrice $\Sigma$ nastale iz dekompozicije originalne slike.

Tako dobijemo žigosanu sliku na kojoj se ne može primijetiti žig.

```python
w=np.matrix(U_orig)*np.matrix(SW)*np.matrix(V_orig)
plt.figure(figsize=(9,6))
plt.imshow(w, cmap='gray')
```

```python
plt.imsave('watermark.png',w)
```

Sada kad smo spremili žigosanu sliku pod nazivom watermark.png možemo je distribuirati. Tijekom distribucije mogu se dogoditi neželjene promijene na slici kao što su rotacija, rezanje(cropping), dodavanje bijelog šuma, zamagljenje(blurring), itd. Dodajmo primjerice zamagljenje, pomoću Gaussovog filtera na našu sliku a nakon toga pokušajmo estrahirati žig.

```python
from scipy.ndimage.filters import gaussian_filter
from scipy.ndimage import *
from skimage.color import rgb2gray

mrkp = imread('watermark.png')
mrkp=rgb2gray(mrkp)
mark_gn=gaussian_filter(mrkp,16)

plt.imsave('watermarkedgn.png',mark_gn)
plt.figure(figsize=(9,6))
plt.imshow(mark_gn);
```

Za rekonstrukciju žiga trebamo originalnu sliku, žig, i žigosanu sliku, te SVD dekompozicije sve tri slike:

```python
U_m,S_M,V_m = svd(markmat)
U_l,S_L,V_l = svd(lenamat)
S_m=diagsvd(S_M,512,512)
S_l=diagsvd(S_L,512,512)
```

```python
D=np.matrix(U_l)*np.matrix(S_m)*np.matrix(V_l)
reconstr = 255*(D - np.matrix(S_don/norm(S_don)))
```
